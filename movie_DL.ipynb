# Movie Genre Classification - Deep Learning Approach
This notebook contains a complete Deep Learning workflow using TensorFlow/Keras to perform multiclass classification on the provided movie dataset.

### **Import the dependencies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
```

### **Load the dataset**

```python
df = pd.read_csv('movie_genre_classification_final.csv')
print('--- First 5 Rows ---')
print(df.head())
print('\n--- Data Shape ---')
print(df.shape)
print('\n--- Data Info ---')
df.info()
```

* Data is **Balanced** across genres.

```python
print(df['Genre'].value_counts())
df = df.drop(columns=['Title'])
df.info()
```

### **Preprocess and Encode Features**

```python
le = LabelEncoder()
df['Genre'] = le.fit_transform(df['Genre'])
df = pd.get_dummies(df, columns=['Description', 'Language', 'Country', 'Production_Company', 'Content_Rating', 'Lead_Actor', 'Director'], drop_first=True)
df.head()
```

### **Split Features and Targets**

```python
x = df.drop(columns='Genre', axis=1)
y = df['Genre']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
print("Training dimensions:", x_train.shape)
```

### **Feature Scaling**

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### **Building the Neural Network**

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(57,)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(7, activation='softmax')
])

model.summary()
```

### **Compile and Train the Network**

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=7, batch_size=32, validation_split=0.1)
```

### **Generate Training Predictions**

```python
train_pred = model.predict(x_train)
train_pred = [np.argmax(i) for i in train_pred]
train_pred
```

### **Generate Testing Predictions**

```python
test_pred = model.predict(x_test)
test_pred = [np.argmax(i) for i in test_pred]
test_pred
```

### **Model Evaluation**

```python
print('Train Acc :-', accuracy_score(y_train, train_pred))
print('Test Acc :-', accuracy_score(y_test, test_pred))

print('\nTrain Confusion Matrix :-\n', confusion_matrix(y_train, train_pred))
print('Test Confusion Matrix :-\n', confusion_matrix(y_test, test_pred))

print('\nTrain Precision :-', precision_score(y_train, train_pred, average='macro'))
print('Test Precision :-', precision_score(y_test, test_pred, average='macro'))

print('\nTrain Recall :-', recall_score(y_train, train_pred, average='macro'))
print('Test Recall :-', recall_score(y_test, test_pred, average='macro'))
```
